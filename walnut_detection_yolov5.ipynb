```python
import subprocess
import re

cmd = [
"python","train.py",
"--img","640",
"--batch","16",
"--epochs","100",
"--data","dataset.yaml",
"--weights","yolov5s.pt",
"--name","baseline_clean",
"--project","/content/drive/MyDrive/walnut_training",
"--workers","0"
]

process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

for line in process.stdout:
    if re.search(r"\s\d+/\d+\s+\d+\.\d+G", line):
        print("✅ Epoch Finished:", line.strip())
```
